```python
import warnings
warnings.filterwarnings('ignore')
prediction=pd.DataFrame()
forecast=pd.DataFrame()
for i in list [0:18]:
    df1=data[['Date','Year','Month',i]]
    if sum((data[i]>0))>=12:
        output=pd.crosstab(columns=df1['Year'],index=df1['Month'],values=df1[i],aggfunc='sum').melt()['value']
        
        output=output.fillna(1)
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
model_fit=ExponentialSmoothing(output,trend='add',seasonal='add',seasonal_periods=12,damped=True)
model_fitted=model_fit.fit(smoothing_level=0,smoothing_trend=0,smoothing_seasonal=0.5,optimized=True)
```

```python
pre=model_fitted.predict(0,47)
```

```python
fore=model_fitted.forecast(3)
```
